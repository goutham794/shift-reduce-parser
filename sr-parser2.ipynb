```python
from ShiftParser import Symbol, Grammar_Rule, Grammar, PNA, PNA_State, PNA_State_Rule

S = Symbol("S", "NonTerminal")
NP = Symbol("NP", "NonTerminal")
VP = Symbol("VP", "NonTerminal")
REL = Symbol("REL", "NonTerminal")
N = Symbol("N", "NonTerminal")
PP = Symbol("PP", "NonTerminal")
V = Symbol("V", "NonTerminal")
PREP = Symbol("PREP", "NonTerminal")
a_cat = Symbol("a_cat", "Terminal")
a_dog = Symbol("a_dog", "Terminal")
a_hat = Symbol("a_hat", "Terminal")
inn = Symbol("inn", "Terminal")
that = Symbol("that", "Terminal")
saw = Symbol("saw", "Terminal")
heard = Symbol("heard", "Terminal")



grammar = Grammar(
    [
    Grammar_Rule(S, [NP,VP]),
    Grammar_Rule(NP, [NP,REL,VP]),
    Grammar_Rule(NP, [N]),
    Grammar_Rule(NP, [N,PP]),
    Grammar_Rule(VP, [V,NP]),
    Grammar_Rule(VP, [V,NP,PP]),
    Grammar_Rule(VP, [V,PP]),
    Grammar_Rule(PP, [PREP,NP]),
    Grammar_Rule(N, [a_cat]),
    Grammar_Rule(N, [a_dog]),
    Grammar_Rule(N, [a_hat]),
    Grammar_Rule(PREP, [inn]),
    Grammar_Rule(REL, [that]),
    Grammar_Rule(V, [saw]),
    Grammar_Rule(V, [heard]),
]
)

pna1 = PNA(grammar)

state_0 = PNA_State([PNA_State_Rule(pna1.grammar.rules[0],0)], pna1)

pna1.construct()
```
