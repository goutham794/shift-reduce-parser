```python
from ShiftParser import Symbol, Grammar_Rule, Grammar, PNA, PNA_State, PNA_State_Rule
```

```python
S = Symbol("S", "NonTerminal")
A = Symbol("A", "NonTerminal")
R = Symbol("R", "NonTerminal")
B = Symbol("B", "NonTerminal")
C = Symbol("C", "NonTerminal")
a = Symbol("a", "Terminal")
b = Symbol("b", "Terminal")
c = Symbol("c", "Terminal")
r = Symbol("r", "Terminal")
```

```python
grammar = Grammar(
    [
    Grammar_Rule(S, [A,R]),
    Grammar_Rule(A, [A,B]),
    Grammar_Rule(A, [a]),
    Grammar_Rule(B, [b]),
    Grammar_Rule(R, [B,R,C]),
    Grammar_Rule(R, [r]),
    Grammar_Rule(C, [c])
]
)
```

```python
pna1 = PNA(grammar)

state_0 = PNA_State([PNA_State_Rule(pna1.grammar.rules[0],0)], pna1)

pna1.construct()
```
